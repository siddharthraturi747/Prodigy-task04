## Sentiment Analysis using Naive Bayes Classifier

#### Preparing the text

```python
df['Feedback'].values
```

```python
# Taking the feedback and sentiments into a seperate dataframe
data = df[['Feedback', 'Sentiment']]
```

#### Converting to lowercase

```python
data['Feedback'] = data['Feedback'].apply(lambda x : x.lower())
```

#### Removing special characters

```python
def remove_special_char(text):
    L = ''
    for i in text:
        if i.isalnum():
            L = L + i
        else:
            L = L + ' '
    return L  
```

```python
data.loc[:,'Feedback'] = data['Feedback'].apply(remove_special_char)
```

#### Stemming

```python
ps = PorterStemmer()

def stem(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
data.loc[:,'Feedback'] = data['Feedback'].apply(stem)
```

```python
data['Feedback'].values
```

```python
# Changing Sentiments into a numerical category
data['Sentiment'] = data['Sentiment'].replace({'Positive':1, 'Negative':0, 'Neutral':2}).astype('int')
data
```

#### Vectorization

```python
cv = CountVectorizer(max_features=15000, stop_words='english')
X = cv.fit_transform(df['Feedback']).toarray()
X
```

```python
y = data.iloc[:,-1]
y
```

## Training the model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

clf2 = MultinomialNB()
```

```python
clf2.fit(X_train, y_train)
```

```python
y_pred2 = clf2.predict(X_test)
```

```python
print('Accuracy : {}'.format(accuracy_score(y_test, y_pred2)))
```

```python
print('Classification report :\n{}'.format(classification_report(y_test, y_pred2)))
```
